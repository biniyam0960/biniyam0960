```python
import json
import nltk
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Download required NLTK models
nltk.download("punkt")
nltk.download("stopwords")

# ==============================
# 1. Load Dataset
# ==============================
json_path = "/content/data/train.json"


with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

print("Loaded dataset:", len(data))

# ==============================
# 2. Build Training Samples
# ==============================
sentences = []
labels = []  # 1 = important, 0 = not important

stop_words = set(stopwords.words("english"))

def preprocess(text):
    words = word_tokenize(text.lower())
    words = [w for w in words if w.isalnum() and w not in stop_words]
    return " ".join(words)

for item in data:
    text = item["text"]
    summary = item["summary"]

    text_sents = sent_tokenize(text)
    summary_sents = sent_tokenize(summary)

    for sent in text_sents:
        sentences.append(preprocess(sent))
        labels.append(1 if any(sent.lower() in s.lower() for s in summary_sents) else 0)

print("Total training sentences:", len(sentences))
print("Label distribution:", labels.count(1), "important,", labels.count(0), "not important")

# Fix issue: If only one class exists → force two classes
if len(set(labels)) == 1:
    print("\n⚠️ Only one class found — adding artificial positive/negative sample.")
    sentences.append("artificial important sentence")
    labels.append(1)
    sentences.append("artificial non important sentence")
    labels.append(0)

# ==============================
# 3. Train Logistic Regression
# ==============================
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
y = np.array(labels)

clf = LogisticRegression(max_iter=800, class_weight="balanced")
clf.fit(X, y)

print("\nModel trained successfully!")

# ==============================
# 4. Summarization Function
# ==============================
def summarize(text, max_sentences=3, max_words=50):

    raw_sents = sent_tokenize(text)
    proc_sents = [preprocess(s) for s in raw_sents]

    # Vectorize
    X_test = vectorizer.transform(proc_sents)
    probs = clf.predict_proba(X_test)[:, 1]

    scored = list(zip(raw_sents, probs))
    scored.sort(key=lambda x: x[1], reverse=True)

    selected = []
    word_count = 0

    # Try to select sentences under constraints
    for sent, score in scored:
        if len(selected) >= max_sentences:
            break
        if word_count + len(sent.split()) <= max_words:
            selected.append(sent)
            word_count += len(sent.split())

    # If NOTHING selected → force pick shortest + top-scoring sentences
    if not selected:
        # Pick top N by probability
        forced = [s for s, _ in scored[:max_sentences]]
        # Then choose the shortest of them
        forced_sorted = sorted(forced, key=lambda x: len(x.split()))
        # Limit words
        final_pick = []
        wcount = 0
        for s in forced_sorted:
            if wcount + len(s.split()) <= max_words or not final_pick:
                final_pick.append(s)
                wcount += len(s.split())
        selected = final_pick

    # Preserve original order
    final = [s for s in raw_sents if s in selected]
    return " ".join(final)

# ==============================
# 5. USER INPUT LOOP
# ==============================
while True:
    text = input("\nEnter text to summarize (or 'exit'): ")

    if text.lower() == "exit":
        break

    num_sents = int(input("How many summary sentences do you want? "))
    num_words = int(input("Max words for summary? "))

    summary = summarize(text, max_sentences=num_sents, max_words=num_words)
    print("\n---------- SUMMARY ----------")
    print(summary)
    print("-----------------------------")
```
